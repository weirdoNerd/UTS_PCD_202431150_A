## Import Library

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Pembacaan Foto Dengan CV2

```python
gambar = cv.imread("FOTO/NAMA.jpeg")
```

## Pembuatan Citra Kontras

```python
kontras = cv.cvtColor(gambar, cv.COLOR_BGR2RGB)
fig, axs = plt.subplots(3, figsize=(8,10))
axs[0].imshow(kontras)
axs[1].hist(kontras.ravel(),256,[0,256])
hist = cv.calcHist([kontras],[0],None,[256],[0,256])
axs[2].plot(hist)

plt.show()
```

## Pembuatan Citra Biru

```python
b = gambar[:, :, 0]

fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(b, cmap='gray')
axs[0,0].set_title("Citra Biru")
axs[0,0].axis('off')

axs[0,1].set_title("Histogram Citra Biru")
axs[0,1].hist(b.ravel(),256,[0,256])

axs[1,0].set_title("Histogram Citra Biru")
hist_blue = cv.calcHist([b],[0],None,[256],[0,256])
axs[1,0].plot(hist_blue)

axs[1,1].hist(kontras.ravel(), bins=256, range=[0,256], color='gray', alpha=0.5, label="Histogram Kontras") 
axs[1,1].hist(b.ravel(), bins=256, range=[0,256], color='blue', alpha=0.5, label="Histogram Biru") 
axs[1,1].set_title("Perbandingan Histogram Citra")
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

## Pembuatan Citra Merah

```python
r = gambar[:, :, 2]

fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(r, cmap='gray')
axs[0,0].set_title("Citra Merah")
axs[0,0].axis('off')

axs[0,1].set_title("Histogram Citra Merah")
axs[0,1].hist(r.ravel(),256,[0,256])

axs[1,0].set_title("Histogram Citra Merah")
hist_red = cv.calcHist([r],[0],None,[256],[0,256])
axs[1,0].plot(hist_red)

axs[1,1].hist(kontras.ravel(), bins=256, range=[0,256], color='gray', alpha=0.5, label="Histogram Kontras") 
axs[1,1].hist(r.ravel(), bins=256, range=[0,256], color='red', alpha=0.5, label="Histogram Merah") 
axs[1,1].set_title("Perbandingan Histogram Citra")
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

## Pembuatan Citra Hijjau

```python
g = gambar[:, :, 1]

fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(g, cmap='gray')
axs[0,0].set_title("Citra Hijau")
axs[0,0].axis('off')

axs[0,1].set_title("Histogram Citra Hijau")
axs[0,1].hist(g.ravel(),256,[0,256])

axs[1,0].set_title("Histogram Citra Hijau")
hist_green = cv.calcHist([b],[0],None,[256],[0,256])
axs[1,0].plot(hist_green)

axs[1,1].hist(kontras.ravel(), bins=256, range=[0,256], color='gray', alpha=0.5, label="Histogram Kontras") 
axs[1,1].hist(g.ravel(), bins=256, range=[0,256], color='green', alpha=0.5, label="Histogram Hijau") 
axs[1,1].set_title("Perbandingan Histogram Citra")
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

## Perbandingan Masing-Masing Citra Biru, Merah, dan Hijau

```python
fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(kontras, cmap='gray')
axs[0,0].set_title("Gambar Asli")
axs[0,0].axis('off')

axs[0,1].imshow(b, cmap='gray')
axs[0,1].set_title("Gambar Biru")
axs[0,1].axis('off')

axs[1,0].imshow(g, cmap='gray')
axs[1,0].set_title("Gambar Hijau")
axs[1,0].axis('off')

axs[1,1].imshow(r, cmap='gray')
axs[1,1].set_title("Gambar Merah")
axs[1,1].axis('off')



plt.tight_layout()
plt.show()
```

```python
b = gambar[:,:,0]
g = gambar[:,:,1]
r = gambar[:,:,2]
```

```python
panjang = len(gambar)
lebar = len(gambar[0])
 
citra_gray = np.zeros((panjang, lebar))
 
for i in range (panjang):
    for j in range (lebar):
        citra_gray[i,j] = 0.3989 * r[i,j] + 0.587 * g[i,j] + 0.1141 * b[i,j]
citra_gray = citra_gray.astype(np.uint8)
```

```python
panjang = len(gambar)
lebar = len(gambar[0])
ambang = 255
gambar_citra = np.zeros((panjang,lebar))
for i in range(panjang):
    for j in range(lebar):
        if citra_gray[i,j] > ambang:
            gambar_citra[i,j] = 1
        else:
            gambar_citra[i,j] = 0
plt.imshow(gambar_citra, cmap='gray')
```

## Nilai Ambang Biru

```python
ambang_bawah_biru = 95
ambang_atas_biru = 255
gambar_b = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        if ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and g[i, j] < 100 and r[i, j] < 100:
            gambar_b[i, j] = 1
        else:
            gambar_b[i, j] = 0
plt.imshow(gambar_b, cmap='gray')
```

##  Nilai Ambang Biru dan Merah

```python
ambang_bawah_merah = 120
ambang_atas_merah = 255
gambar_m = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        biru = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        merah = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        if merah or biru:
            gambar_m[i, j] = 1
        else:
            gambar_m[i, j] = 0 

plt.imshow(gambar_m, cmap='gray')
```

## Nilai Ambang Semua Warna

```python
ambang_bawah_hijau = 95
ambang_atas_hijau = 255
gambar_bgr = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        biru = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        hijau = ambang_bawah_hijau <= g[i, j] <= ambang_atas_hijau and r[i, j] < 100 and b[i, j] < 100
        merah = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        if merah or biru or hijau:
            gambar_bgr[i, j] = 1
        else:
            gambar_bgr[i, j] = 0 

plt.imshow(gambar_bgr, cmap='gray')
```

## Perbandingan Nilai Ambang Warna

```python
fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(gambar_citra, cmap='gray')
axs[0,0].set_title("Ambang None")
axs[0,0].axis('off')

axs[0,1].set_title("Ambang Biru")
axs[0,1].imshow(gambar_b, cmap='gray')

axs[1,0].set_title("Ambang Biru dan Merah")
axs[1,0].imshow(gambar_m, cmap='gray')

axs[1,1].set_title("Ambang Biru, Hijau, dan Merah")
axs[1,1].imshow(gambar_bgr, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
(baris, kolom) = img.shape[:2]
```

## Foto Asli

```python
img_asli = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_asli)
plt.axis("off")
```

## Foto Abu

```python
img_abu = cv.cvtColor(img_asli, cv.COLOR_BGR2GRAY)
plt.imshow(img_abu, cmap='gray')
plt.axis('off')
```

## Mencerahkan Foto

```python
beta = 17
img_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = img_abu[x,y] + beta
        img_cerah[x,y] = gyx

img_cerah = img_cerah.astype(np.uint8)

fig, axs = plt.subplots(2, figsize=(20,10))
axs[0].imshow(img_abu, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img_cerah)
axs[1].axis('off')
plt.show()
```

## Kontras Foto

```python
alpha = 1.05
img_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gmx = img_abu[x,y] * alpha
        img_kontras[x,y] = gmx

img_kontras = img_kontras.astype(np.uint8)

fig, axs = plt.subplots(2, figsize=(20,10))
axs[0].imshow(img_abu, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img_kontras)
axs[1].axis('off')
plt.show()
```
